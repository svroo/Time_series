1. ¿Cómo ha evolucionado la calidad del aire a lo largo del tiempo en Myanmar? ¿Existen patrones de estacionalidad, tendencias o ciclos en los niveles de contaminación?

```python
from sklearn.linear_model import LinearRegression

# Seleccionamos valores de la ciudad mandaly
mandalay = features[features.City == 'Mandalay']

# Convertir la columna de fechas al formato datetime
# mandalay['Date'] = pd.to_datetime(mandalay['Date'], format='%m/%d/%Y hh:mm')
mandalay['Date'] =pd.to_datetime(mandalay['Date'], format='%m/%d/%Y %H:%M')

# Establecer la fecha como índice
mandalay.set_index('Date', inplace=True)
aqi =features['AQI'][features.City == 'Mandalay']

# Graficar la serie temporal de los niveles de contaminación
plt.figure(figsize=(12,6))
# plt.plot(mandalay['AQI'])
plt.plot(features['AQI'][features.City == 'Mandalay'])
plt.title('Evolución de la calidad del aire en Myanmar')
plt.xlabel('Fecha')
plt.ylabel('AQI')
plt.show()

# Resamplear los datos a una frecuencia mensual
df_monthly = mandalay.resample('M').mean()

# Graficar la serie temporal resampleada
plt.figure(figsize=(12,6))
plt.plot(df_monthly['AQI'])
plt.title('Evolución mensual de la calidad del aire en Myanmar')
plt.xlabel('Fecha')
plt.ylabel('AQI')
plt.show()

# Calcular la media móvil de 12 meses
rolling_mean = df_monthly.rolling(window=12, min_periods=1).mean()

# Graficar la serie temporal resampleada junto con la media móvil
plt.figure(figsize=(12,6))
plt.plot(df_monthly['AQI'], label='AQI')
plt.plot(rolling_mean['AQI'], label='Media móvil de 12 meses')
plt.title('Evolución mensual de la calidad del aire en Myanmar con media móvil de 12 meses')
plt.xlabel('Fecha')
plt.ylabel('AQI')
plt.legend()
plt.show()

# Calcular la regresión lineal de los niveles de contaminación a lo largo del tiempo
def linear_regression(df: pd.DataFrame) -> Tuple[float, float]:
    X = np.arange(len(df)).reshape(-1,1)
    y = df['AQI'].values.reshape(-1,1)
    reg = LinearRegression().fit(X, y)
    return reg.coef_[0][0], reg.intercept_[0]

slope, intercept = linear_regression(df_monthly)
t = np.arange(len(df_monthly))
trendline = intercept + slope * t

# Graficar la serie temporal resampleada junto con la media móvil y la línea de tendencia
plt.figure(figsize=(12,6))
plt.plot(df_monthly['AQI'], label='AQI')
plt.plot(rolling_mean['AQI'], label='Media móvil de 12 meses')
plt.plot(trendline, label='Línea de tendencia')
plt.title('Evolución mensual de la calidad del aire en Myanmar con media móvil de 12 meses y línea de tendencia')
plt.xlabel('Fecha')
plt.ylabel('AQI')
plt.legend()
plt.show()

# Calcular la estacionalidad de los niveles de contaminación a lo largo del año
df_yearly = mandalay.resample('Y').mean()
df_yearly.reset_index(inplace=True)
df_yearly['year'] = df_yearly['Date'].dt.year
df_yearly.drop(columns=['Date'], inplace=True)
df_yearly.set_index('year', inplace=True)

sns.lineplot(data=df_yearly, x=df_yearly.index, y='AQI')
plt.title('Patrón estacional de la calidad del aire en Myanmar')
plt.xlabel('Año')
plt.ylabel('Índice de calidad del aire (AQI)')
```

¿Cómo se han visto afectados los niveles de contaminación por la pandemia de COVID-19? 

```python
# Limpiar y preprocesar los datos
copia = features
copia['Date'] = pd.to_datetime(copia['Date'])
copia = copia.set_index('Date')

# Paso 2: Analizar los datos
# Calcular estadísticas descriptivas de los niveles de contaminación en cada ciudad, tanto antes como después de la pandemia de COVID-19
before_covid = copia[copia.index < '2020-03-01'].groupby('City')['AQI'].mean()
after_covid = copia[copia.index >= '2020-03-01'].groupby('City')['AQI'].mean()

# Graficar los datos para visualizar cualquier patrón o tendencia en los niveles de contaminación a lo largo del tiempo y entre las distintas ciudades
sns.set_style('darkgrid')
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
sns.boxplot(data=copia, x='City', y='AQI', ax=axs[0])
axs[0].set_title('Niveles de contaminación en cada ciudad')
sns.barplot(data=pd.DataFrame({'Antes de COVID-19': before_covid, 'Después de COVID-19': after_covid}).reset_index().melt(id_vars='City'), x='City', y='value', hue='variable', ax=axs[1])
axs[1].set_title('Comparación de niveles de contaminación antes y después de COVID-19')
plt.tight_layout()
plt.show()
```

¿Hay alguna diferencia en la evolución de los niveles de contaminación entre las distintas ciudades de Myanmar?

```python
aux = features

# Obtenemos la columna como espacio de tiempo
aux['Date'] = pd.to_datetime(aux.Date)
aux['year'] = aux.Date.dt.year
aux_by_city_year = aux.groupby(['City', 'year'])['AQI'].mean().reset_index()

# Crear una gráfica que muestre la evolución de los niveles de contaminación por ciudad
sns.lineplot(data=aux_by_city_year, x='year', y='AQI', hue='City')
plt.title('Evolución de los niveles de contaminación por ciudad en Myanmar')
plt.xlabel('Año')
plt.ylabel('Índice de Calidad del Aire')
plt.show()
```
